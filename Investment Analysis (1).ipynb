# Simulate Holding Position Amaount in Amazon
We will hold 10 stock in amazon

```python
initial_position =10000
```

```python
amzn["percent_change_shift"] = amzn["percent_change"].shift(-1)       # shift the percent change columnn up 1
amzn["percent_change_shift"] = amzn["percent_change_shift"].fillna(0) # then change the NA at end to be 0
amzn
```

```python
# Simulate the current holding position of your portfolio now..

new = [(amzn.percent_change_shift.values[0] * initial_position) + initial_position] # make the first value of new column hold it in variable where we take initial position and multiply by the shifted percent change column to get initial starting value
for i in range(1, len(amzn.index)): # starting at second row of the dataframe column 
    new.append((new[i-1]*amzn.percent_change_shift.values[i]) +new[i-1]) # take new and multiply the first calculated value by the next percent change plus the origional value to get new position value

# Assign the list new to the dataframe column
amzn['CurrentPosition'] = new
amzn['CurrentPosition'] = amzn['CurrentPosition'].round(2) # round the coumn
amzn['Current_Position_Final'] = amzn['CurrentPosition'].round(2).shift(1) # shift it down to make since since shifted percent change up one value
amzn.Current_Position_Final.iloc[0] = 10000  # inject the initial starting amount into the first value of the column.. 
amzn
```

```python
# We can see if we would have had 10k in amazon we would now hav 15k in amazon... 

amzn2 = amzn[["adjclose","percent_change","Current_Position_Final"]]
amzn2
```

```python
# Make unique function to add payment at end of month 

# amzn2.reset_index(level=0, inplace=True)
amzn2
amzn2["formatted_date"] =  pd.to_datetime(amzn2.formatted_date)
amzn2['day'] = pd.DatetimeIndex(amzn2['formatted_date']).day
amzn2['month'] = pd.DatetimeIndex(amzn2['formatted_date']).month
amzn2['year'] = pd.DatetimeIndex(amzn2['formatted_date']).year

amzn2
```

```python
# This gets the max date for each month in python and only keeps those columns 
max_date_of_each_month = amzn2.resample(on="formatted_date", rule="M").max().reset_index(drop=True)

# maek it dataframe 
max_date_of_each_month = pd.DataFrame(max_date_of_each_month)

# keep only these columns
max_date_of_each_month = max_date_of_each_month[["formatted_date","day"]]

# make this so when left join if 1 then add the increment amount
max_date_of_each_month["end_of_month"] = 1

#del max_date_of_each_month.day
max_date_of_each_month
```

```python
# left join the end of month values to the same data frame 
amzn3 = amzn2.merge(max_date_of_each_month, on='formatted_date', how='left')
amzn3
```
