# Financial Analysis

```python
# Daily % change or (daily return)
#rt = pt/pt-1 -1 
```

```python
start = datetime.datetime(2012,1,1)
end = datetime.datetime(2021,1,1)
tesla = web.DataReader('TSLA','yahoo',start,end)
ford = web.DataReader('F','yahoo',start,end)
GM = web.DataReader('GM','yahoo',start,end)
```

```python
# Get daily returns THEY ARE BOTH THE SAME
tesla['returns'] = (tesla['Close']/tesla['Close'].shift(1)) - 1
tesla['returns'] = tesla['Close'].pct_change(1) # one time period or one day
tesla['returns']
```

```python
GM['returns'] = GM['Close'].pct_change(1) # one time period or one day
ford['returns'] = ford['Close'].pct_change(1) # one time period or one day
GM.head()
```

```python
ford['returns'].hist(bins=100) # get histograme of returns bin 10 by default
```

```python
GM['returns'].hist(bins=100)
```

```python
tesla['returns'].hist(bins=100,)
```

```python
# Introduce alpha to see through 
tesla['returns'].hist(bins=100,label='Tesla',figsize=(10,8))
ford['returns'].hist(bins=100,label='ford',figsize=(10,8),alpha =0.5)
GM['returns'].hist(bins=100,label='GM',figsize=(10,8),alpha =0.5)
plt.legend()
```

# Kernel density

```python
tesla['returns'].plot(kind='kde',label='Tesla',figsize=(10,8))
ford['returns'].plot(kind='kde',label='ford',figsize=(10,8))
GM['returns'].plot(kind='kde',label='GM',figsize=(10,8))
plt.legend()
# ford is more stable
```

# Comparing returns

```python
box_df = pd.concat([tesla['returns'],ford['returns'],GM['returns']],axis=1)
box_df.columns = ['Tesla Ret','Ford Ret','GM Ret']
box_df.plot(kind='box',figsize=(8,11))
```

# Daily returns between stocks 

```python
scatter_matrix(box_df,figsize=(8,8),alpha = 0.2,hist_kwds={'bins':100});
```

```python
# ford and gm might have relationship so lets look closer
box_df.plot(kind='scatter',x='Ford Ret',y= 'GM Ret', alpha = 0.5,figsize=(8,8))
```

# Cumulative Daily Returns

```python
# Daily returns doesnt give insight to how much money you improved on 
# if cumulative return is above 1 your good else under one bad
tesla['Cumulative Return'] = (1 + tesla['returns']).cumprod()
```

```python
GM['Cumulative Return'] = (1 + GM['returns']).cumprod()
ford['Cumulative Return'] = (1 + ford['returns']).cumprod()
```

```python
# Plot cumulative returns against time series 
tesla['Cumulative Return'].plot(label='Tesla',figsize=(16,8))
ford['Cumulative Return'].plot(label='ford',figsize=(16,8))
GM['Cumulative Return'].plot(label='GM',figsize=(16,8))
plt.legend()
```
